```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import LineCollection
from matplotlib.colors import Normalize

# 🔹 1. Load your CSV file
df = pd.read_csv("/Users/tuba/Downloads/UMB_Boston/lee/CORAL/CGU1704/combined_rrs.csv")

# Ensure proper column names
df.columns = df.columns.str.strip().str.lower()  # make lowercase to match "rrs"
print(df.head())

# 🔹 2. Convert to numeric
df["wavelength"] = pd.to_numeric(df["wavelength"], errors="coerce")
df["rrs"] = pd.to_numeric(df["rrs"], errors="coerce")
df = df.dropna(subset=["wavelength", "rrs"])

# 🔹 3. Prepare points for colored line segments
points = np.array([df["wavelength"], df["rrs"]]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

# 🔹 4. Create a colormap based on wavelength
norm = Normalize(df["wavelength"].min(), df["wavelength"].max())
lc = LineCollection(segments, cmap="turbo", norm=norm)
lc.set_array(df["wavelength"])
lc.set_linewidth(2)

# 🔹 5. Plot
fig, ax = plt.subplots(figsize=(10, 6))
line = ax.add_collection(lc)

ax.set_xlim(df["wavelength"].min(), df["wavelength"].max())
ax.set_ylim(df["rrs"].min(), df["rrs"].max())

plt.xlabel("Wavelength (nm)", fontsize=12)
plt.ylabel("Rrs (sr⁻¹)", fontsize=12)
plt.title("Spectral Reflectance (Rrs vs Wavelength)", fontsize=14, fontweight="bold")
plt.grid(True, alpha=0.3)

# 🔹 Add a colorbar showing wavelength
cbar = plt.colorbar(line, ax=ax)
cbar.set_label("Wavelength (nm)")

plt.tight_layout()
plt.show()
```
